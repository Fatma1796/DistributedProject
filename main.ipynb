```python
from collections import defaultdict

# Correct file path
#file_path = r"C:\Users\My Lab\Desktop\Distributed project\DistributedProject\coursegrades.txt"
file_path = r"C:\Users\ganna\PycharmProjects\DistributedProject\coursegrades.txt"

# TASK 1: Average Grade per Course
# (course name, grade) mapper
def mapper_t1(file_path):
    mapped_data = []
    with open(file_path, "r") as file:
        for line in file:
            line = line.strip()
            parts = line.split(",")

            if len(parts) < 4:
                continue

            course = parts[1].strip()
            grade = parts[2].strip()

            mapped_data.append((course, int(grade)))

    return mapped_data

def reducer_t1(mapped_data):
    grades_by_course = defaultdict(list)

    for course, grade in mapped_data:
        grades_by_course[course].append(grade)

    averaged_grades = {}
    for course, grades in grades_by_course.items():
        average_grade = sum(grades) / len(grades)
        averaged_grades[course] = average_grade

    highest_avg_course = max(averaged_grades, key=averaged_grades.get)
    highest_avg = averaged_grades[highest_avg_course]

    return averaged_grades, (highest_avg_course, highest_avg)

#run mapper and the output is passes to the reducer
mapper_output_T1 = mapper_t1(file_path)
averages, highest_course = reducer_t1(mapper_output_T1)

print("Average Grades per Course:")
for course, avg_grade in averages.items():
    print(f"{course}: {avg_grade:.2f}")

print("\nCourse with the Highest Average:")
print(f"{highest_course[0]}: {highest_course[1]:.2f}")

print()

# bonus task: top 3 highest grades per year
# Mapper function to process the file
def mapper(file_path):
    mapped_data = []  # Store the output instead of printing it
    with open(file_path, "r") as file:
        for line in file:
            line = line.strip()
            parts = line.split(",")  # Split by comma

            if len(parts) < 4:  # Ensure correct format
                continue

            year = parts[0].strip()   # Extract year
            grade = parts[2].strip()  # Extract grade

            mapped_data.append(f"{year}\t{grade}")  # Store instead of print

    return mapped_data  # Return the mapped data for next step

# Reducer function to find the top 3 highest grades per year
def reducer(mapped_data):
    grades_by_year = defaultdict(list)

    # Process mapped data
    for line in mapped_data:
        year, grade = line.strip().split("\t")  # Read mapper output
        grades_by_year[year].append(int(grade))  # Store grades by year

    # Compute top 3 grades for each year
    result = {}
    for year, grades in grades_by_year.items():
        top_3 = sorted(grades, reverse=True)[:3]  # Sort in descending order
        result[year] = top_3  # Store results

    return result

# Run the mapper function and pass its output to the reducer
mapper_output = mapper(file_path)
top_3_grades = reducer(mapper_output)

# Print the results
print("Top 3 Grades per Year:")
for year, grades in top_3_grades.items():
    print(f"{year}: {grades}")

```
